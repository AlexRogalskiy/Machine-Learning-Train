```python
X = iris.data[:,:2]
X.shape
```

```python
y = iris.target
```

```python
plt.scatter(X[y==0,0],X[y==0,1],color="red")
plt.scatter(X[y==1,0],X[y==1,1],color="blue")
plt.scatter(X[y==2,0],X[y==2,1],color="green")
plt.show()
```
